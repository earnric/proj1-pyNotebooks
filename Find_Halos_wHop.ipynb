```python
import sys
import numpy as np
from subprocess import Popen, PIPE # if we want to get the output.
import subprocess
from multiprocessing import Pool, Lock
import time
```

```python
filesToProcess = [
    "00008",
    "00016",
    "00026",
    "00043",
    "00058",
    "00073",
    "00097",
    "00121",
    "00152",
    "00191" ]
```

```python
hopitProc = np.empty(len(filesToProcess),dtype=subprocess.Popen)
for i, aFile in enumerate(filesToProcess):
    print("Calling hopit: output_%s"%aFile)
    hopitProc[i] = subprocess.Popen(['./hopit',aFile], stdin=subprocess.PIPE, stdout=subprocess.PIPE, 
                                    stderr=subprocess.PIPE)
#     for line in iter(hopit.stdout.readline, ''):
#         sys.stdout.write(line)
#         #f.write(line)
```

```python
loop=True;
while loop is True:
    for i, proc in enumerate(hopitProc):
        time.sleep(20)
        status = proc.poll()
        if status == None:
            print("process %d running..."%i)
            continue
        elif status == 0:
            print("process %d comleted"%i)
            loop=False
        else:
            print("process %d failed with status"%i, status)
            loop=False
```
